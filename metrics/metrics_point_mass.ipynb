```python
%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.4
n = 10

for i in range(n):
    path = '../logs/point_mass/robot_' + str(i + 1) + '.txt'

    with open(path, "r") as file:
        plt.title("x(t), y(t), z(t) for robot_" + str(i + 1))
        plt.xlabel('t')
        plt.ylabel('x, y, z')
        t = 0

        ts, xs, ys, zs = [], [], [], []

        for line in file:
                nums = [float(item) for item in line.rstrip().split(' ')]
                x, y, z, dx, dy, dz, d_x_plus, d_x_minus, d_y_plus, d_y_minus, d_z_plus, d_z_minus = nums

                ts.append(t)
                xs.append(x)
                ys.append(y)
                zs.append(z)

                t += dt

        plt.plot(ts, xs, color='blue')
        plt.plot(ts, ys, color='red')
        plt.plot(ts, zs, color='green')

        plt.show()
```

```python
for i in range(n):
    path = '../logs/point_mass/robot_' + str(i + 1) + '.txt'

    with open(path, "r") as file:
        plt.title("ux(t), uy(t), uz(t) for robot_" + str(i + 1))
        plt.xlabel('t')
        plt.ylabel('ux, uy, uz')
        t = 0

        ts, xs, ys, zs = [], [], [], []

        for line in file:
                nums = [float(item) for item in line.rstrip().split(' ')]
                x, y, z, dx, dy, dz, d_x_plus, d_x_minus, d_y_plus, d_y_minus, d_z_plus, d_z_minus = nums

                ts.append(t)
                xs.append(dx)
                ys.append(dy)
                zs.append(dz)

                t += dt

        plt.plot(ts, xs, color='blue')
        plt.plot(ts, ys, color='red')
        plt.plot(ts, zs, color='green')

        plt.show()
```

```python
for i in range(n):
    path = '../logs/point_mass/robot_' + str(i + 1) + '.txt'

    with open(path, "r") as file:
        plt.title("d_x_plus and d_x_minus for robot_" + str(i + 1))
        plt.xlabel('t')
        plt.ylabel('d_x_plus, d_x_minus')
        t = 0

        ts, dxs_plus, dxs_minus, diffs = [], [], [], []

        for line in file:
                nums = [float(item) for item in line.rstrip().split(' ')]
                x, y, z, dx, dy, dz, d_x_plus, d_x_minus, d_y_plus, d_y_minus, d_z_plus, d_z_minus = nums

                ts.append(t)
                dxs_plus.append(d_x_plus)
                dxs_minus.append(-d_x_minus)
                diffs.append(d_x_plus - d_x_minus)

                t += dt

        plt.plot(ts, dxs_plus, color='blue')
        plt.plot(ts, dxs_minus, color='red')
        plt.plot(ts, diffs, color='black')

        plt.show()
```

```python
colors = ['red', 'green', 'blue', 'orange', 'aqua', 'magenta', 'black', 'purple', 'grey', 'brown']

for i in range(n):
    path = '../logs/point_mass/robot_' + str(i + 1) + '.txt'

    with open(path, "r") as file:
        xs, zs = [], []

        for line in file:
                nums = [float(item) for item in line.rstrip().split(' ')]
                x, y, z, dx, dy, dz, d_x_plus, d_x_minus, d_y_plus, d_y_minus, d_z_plus, d_z_minus = nums

                xs.append(x)
                zs.append(z)

        plt.plot(xs, zs, color=colors[i])

plt.xlabel('x')
plt.ylabel('z')
plt.title('XZ projections of robots trajectories')
plt.show()
```

```python
for i in range(n):
    path = '../logs/point_mass/robot_' + str(i + 1) + '.txt'

    with open(path, "r") as file:
        ts, ds = [], []
        t = 0

        for line in file:
                nums = [float(item) for item in line.rstrip().split(' ')]
                x, y, z, dx, dy, dz, d_x_plus, d_x_minus, d_y_plus, d_y_minus, d_z_plus, d_z_minus = nums
                d = np.sqrt(x**2 + y**2)

                ts.append(t)
                ds.append(d)

                t += dt

        plt.plot(ts, ds, color=colors[i])

plt.title('Distanses from target axis in XY-projection')
plt.xlabel('t')
plt.ylabel('d')
plt.show()
```
