```python
%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.008
path = '../logs/test/log_1.txt'

with open(path, "r") as file:
    plt.title("Pitch disturbance control and real pitch")
    plt.xlabel('t')
    plt.ylabel('rad')
    t = 0

    ts, pitches, pitch_ds, upitches = [], [], [], []

    for line in file:
            nums = [float(item) for item in line.rstrip().split(' ')]
            roll, pitch, yaw, uroll, upitch, uyaw, roll_dist, pitch_dist, yaw_dist = nums

            ts.append(t)
            pitches.append(pitch)
            pitch_ds.append(pitch_dist)
            upitches.append(upitch)

            t += dt

    plt.plot(ts, pitches, color='blue')
    plt.plot(ts, pitch_ds, color='green')

    plt.show()
```
