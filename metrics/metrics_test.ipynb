```python
#%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.008
path = '../logs/test/log_1.txt'

with open(path, "r") as file:
    plt.title("Pitch disturbance control and real pitch")
    plt.xlabel('t')
    plt.ylabel('rad')
    t = 0

    ts, pitches, pitch_ds, upitches = [], [], [], []

    for line in file:
            nums = [float(item) for item in line.rstrip().split(' ')]
            roll, pitch, yaw, uroll, upitch, uyaw, roll_dist, pitch_dist, yaw_dist = nums

            ts.append(t)
            pitches.append(pitch)
            pitch_ds.append(pitch_dist)
            upitches.append(upitch)

            t += dt

    plt.plot(ts, pitches, color='blue', label=r"$\theta$")
    plt.plot(ts, pitch_ds, color='green', label=r"$\Delta\theta$")
    plt.legend(fontsize=10)

    plt.show()
```

```python
with open(path, "r") as file:
    plt.title("Pitch and roll")
    plt.xlabel('t')
    plt.ylabel('rad')
    t = 0

    ts, pitches, upitches, rolls, urolls, yaws, uyaws = [], [], [], [], [], [], []

    for line in file:
            nums = [float(item) for item in line.rstrip().split(' ')]
            roll, pitch, yaw, uroll, upitch, uyaw, roll_dist, pitch_dist, yaw_dist = nums

            ts.append(t)
            pitches.append(pitch)
            upitches.append(upitch/20)
            rolls.append(roll)
            urolls.append(uroll)
            yaws.append(yaw)
            uyaws.append(uyaw)

            t += dt

    plt.plot(ts, pitches, color='blue', label=r"$\theta$")
    plt.plot(ts, rolls, color='green', label=r"$\phi$")
    plt.legend(fontsize=10)

    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigma(x):
    return np.tanh(x)

def clinear(x):
    if x > 1:
        return 1
    if x < -1:
        return -1
    return x

xs = np.linspace(-2, 2, 100)
us = [clinear(x) for x in xs]

plt.plot(xs, us)
plt.xlabel('e')
plt.ylabel('u')
plt.grid(which='minor', linestyle=':')

plt.show()
```
